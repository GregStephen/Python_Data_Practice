```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

CATEGORICAL PLOTS

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split='True')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue="sex", dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

MATRIX PLOTS

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fl = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fl, cmap='magma', linecolor='silver' , linewidths=2)
```

```python
sns.clustermap(fl, cmap='coolwarm', standard_scale=1)
```

GRIDS

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tg = sns.FacetGrid(data=tips, col='time', row='smoker')
tg.map(sns.distplot, 'total_bill')
```
